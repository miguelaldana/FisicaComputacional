```python
n_random = 10000

min_val = -1.0
max_val = 1.0

random_x = random.rand(n_random) * (max_val - min_val) + min_val
random_y = random.rand(n_random) * (max_val - min_val) + min_val

delta = random_x**2 + random_y**2

inside = where(delta < 1.0)

scatter(random_x[inside], random_y[inside])
```

```python
interval_integral = (max_val-min_val) * (max_val - min_val)
integral  = interval_integral * (size(inside)/(1.0*size(random_y)))

print integral
```
