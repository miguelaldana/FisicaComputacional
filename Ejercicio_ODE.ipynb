```python
h = 0.1
n_points = int((1.0 + h)/h)
x = zeros(n_points)
y = zeros(n_points)
```

```python
def func_prime(x,y):
    return 1 + y**2
```

```python
x[0] = 0.0
y[0] = 0.0
for i in range(1,n_points):
    
    k1 = func_prime(x[i-1],y[i-1])
    
    x1 = x[i-1] + h/2
    y1 = y[i-1] + (h/2)*k1
    k2 = func_prime(x1,y1)
    
    x2 = x[i-1] + h/2
    y2 = y[i-1] + (h/2)*k2
    k3 = func_prime(x2,y2)
    
    x3 = x[i-1] + h
    y3 = y[i-1] + h*k3
    k4 = func_prime(x3,y3)
    
    average = (k1 + 2*k2 + 2*k3 + k4)/6
    x[i] = x[i-1] + h
    y[i] = y[i-1] + h*average
```

```python
plot(x,y, 'ko')
plot(x, tan(x))
```

```python
desv = zeros(n_points)
for i in range(n_points):
    desv[i] = tan(x[i])-y[i]
```

```python
plot(x, abs(tan(x)-y))
```

```python
h = 0.05
n_points = int((6+h)/h)
x = zeros(n_points)
y_1 = zeros(n_points)
y_2 = zeros(n_points)
```

```python
x[0] = 0.0
y_1[0] = 0.1
y_2[0] = 0.0

def func_prime_1(x,y_1,y_2):
    return y_2

def func_prime_2(x,y_1,y_2):
    return -4*sin(y_1)

for i in range(1,n_points):
    
    k1 = func_prime_1(x[i-1],y_1[i-1],y_2[i-1])
    k2 = func_prime_2(x[i-1],y_1[i-1],y_2[i-1])
    
    x1 = x[i-1] + h/2
    y1 = y_1[i-1] + (h/2)*k1
    y__2 = y_2[i-1] + (h/2)*k2
    k3 = func_prime_1(x1,y1,y__2)
    k4 = func_prime_2(x1,y1,y__2)
    
    x2 = x[i-1] + h/2
    y2 = y_1[i-1] + (h/2)*k3
    y__3 = y_2[i-1] + (h/2)*k4
    k5 = func_prime_1(x2,y2,y__3)
    k6 = func_prime_2(x2,y2,y__3)
    
    x3 = x[i-1] + h
    y3 = y_1[i-1] + (h)*k5
    y__4 = y_2[i-1] + (h)*k6
    k7 = func_prime_1(x3,y3,y__4)
    k8 = func_prime_2(x3,y3,y__4)
    
    average_1 = (k1 + 2*k3 + 2*k5 + k7)/6
    average_2 = (k2 + 2*k4 + 2*k6 + k8)/6
    
    x[i] = x[i-1] + h
    y_1[i] = y_1[i-1] + h*average_1
    y_2[i] = y_2[i-1] + h*average_2
```

```python
plot(x,y_1, 'ko')
plot(x,0.1*cos(2.0*x))
plt.xlabel('x')
plt.ylabel('y(x)')
```
