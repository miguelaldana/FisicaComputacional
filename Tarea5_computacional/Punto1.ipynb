```python
%pylab inlineimport numpy as np
```

```python
data = np.loadtxt("sampled+ma0844az_1-1+_data.txt")
```

```python
from scipy.fftpack import fft, fftfreq, ifftimport matplotlib.pyplot as plt
```

```python
n, col = shape(data)matr = zeros((n,col))for i in range(24):    matr[:,i] = data[:,i]plt.plot(matr)
```

```python
fft_amplitudes = empty([n,col], dtype=complex)for j in range(24):    fft_amplitudes[:,j] = fft(data[:,j])freq = fftfreq(n,1)plt.plot(freq, np.abs(fft_amplitudes)**2)
```

```python
half_n = np.ceil(n/2.0)fft_amplitudes_half = empty([n/2,col], dtype=complex)freq_half = freq[:half_n]for i in range(24):    fft_amplitudes_half[:,i] = fft_amplitudes[:half_n,i]    plt.plot(freq_half, np.abs(fft_amplitudes_half[:,i])**2.0)
```

```python
max_espectro_x = empty((n,col))for i in range(24):    max_espectro_x[:,i] = np.sort(np.abs(fft_amplitudes[:,i]), axis=None)    max_espectro_x[:,i] = max_espectro_x[::-1,i]for j in range(24):    for i in range(n):        if i > 9:            max_espectro_x[i,j] = 0.0for j in range(24):    for u in range(n):        for i in range(n):            if max_espectro_x[u,j] != 0.0:                if np.abs(fft_amplitudes[i,j]) == max_espectro_x[u,j]:                    continue                else:                    fft_amplitudes[i,j] = 0.0plt.plot(freq, np.abs(fft_amplitudes[:,0])**2, 'r')
```
