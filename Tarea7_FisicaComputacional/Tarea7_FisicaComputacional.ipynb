```python
n_x = 200
n_t = 500
T = 0.15

c = 1.0
nu = 0.7 #valor para asegurar que beta (dt/dx) es menor a uno

x = linspace(0, 2.0, n_x)
dx = x[1]-x[0]


dt = T/n_t
beta = c/(dx/dt)

print c
print dx
print dt
print beta

#Condiciones iniciales del problema

u = 3.0*sin(pi*x)
u[0] = u[1] = 0.0

plot(x,u)
```

```python
#solucionamos por medio de la ec. 2
import copy

u_viejo = u.copy()
u_nuevo = u.copy()
u_nuevo_2 = zeros(n_x)
u_futuro = zeros(n_x)
u_presente = zeros(n_x)
u_total = empty([n_x,n_t])

for j in range(n_t):
    
    for i in range(1,n_x-1):
        
        u_presente[i] = ((u_nuevo[i+1]**2) - (u_nuevo[i-1]**2))/2
        u_nuevo_2[i] = beta*u_presente[i]
        u_futuro[i] = u_viejo[i] - u_nuevo_2[i]
        
    u_viejo = u_nuevo.copy()
    
    u_nuevo = u_futuro.copy()
    
    u_total[:,j] = u_nuevo[:]
    
plt.figure(figsize=(11,7), dpi=100)
xlim(0,2.0)
    
plot(x,u_nuevo)
plot(x, u_total)

r = 0
```

```python
#solucionamos por medio del metodo de Lax-Wendroff

u_presente_2 = u.copy()
u_temporal_1 = zeros(n_x)
u_temporal_2 = zeros(n_x)
u_temporal_3 = zeros(n_x)
u_futuro_2 = zeros(n_x)
u_total_2 = empty([n_x,n_t])

for j in range(n_t):
    for i in range(1,n_x-1):
        u_temporal_1[i] = (u_presente_2[i] + u_presente_2[i-1])*((u_presente_2[i]**2) - (u_presente_2[i-1]**2))
        u_temporal_2[i] = (u_presente_2[i+1] + u_presente_2[i])*((u_presente_2[i+1]**2) - (u_presente_2[i]**2))
        u_temporal_3[i] = ((u_presente_2[i+1]**2) - (u_presente_2[i-1]**2))
        u_futuro_2[i] = u_presente_2[i] - (beta/4.0)*u_temporal_3[i] + ((beta**2)/8.0)*(u_temporal_2[i] - u_temporal_1[i])
        
    u_presente_2 = u_futuro_2.copy()
    
    u_total_2[:,j] = u_presente_2[:]

plt.figure(figsize=(11,7), dpi=100)
x = linspace(0, 2.0, n_x)
xlim(0,2.0)
    
plot(x,u_presente_2)
plot(x, u_total_2)

r = 0
```
